# Model cluster dengan k=2

```python
from scipy.cluster.vq import whiten, kmeans, vq 
skala_data2 = whiten(data) 
print(skala_data2)
```

```python
skala_data2 = pd.DataFrame(skala_data2, columns=['Alcohol', 'Malic_Acid', 'Ash', 'Ash_Alcanity','Magnesium', 'Total_Phenols', 'Flavanoids', 'Nonflavanoid_Phenols', 'Proanthocyanins', 'Color_Intensity', 'Hue', 'OD280', 'Proline', 'Customer_Segment'])
cluster_centers, distortion = kmeans(skala_data2, 2) 
skala_data2['label'], distortions_list = vq(skala_data2, cluster_centers)
```

```python
labels2 = skala_data2['label'] 
skala2 = skala_data2.drop('label', axis=1) 
print(skala2.columns) 
print(labels2)
```

```python
from sklearn.decomposition import PCA 
model2 = PCA(n_components=2) 
model2.fit(skala2) 
X = model2.transform(skala2)
```

```python
X_final2 = pd.DataFrame(X, columns=['X', 'Y']) 
X_final2.head()
```

```python
labels2 = pd.DataFrame(labels2, columns=['label']) 
dfs = [X_final2, labels2] 
final_df2 = pd.concat(dfs, axis=1) 
final_df2.head()
```

```python
plt.figure(figsize=(8, 8)) 
sns.scatterplot(x='X', y='Y', hue='label', data=final_df2);
```

# model cluster dengan menggunakan metode elbow

```python
from scipy.cluster.vq import whiten, kmeans, vq 
skala_data = whiten(data) 
print(skala_data)
```

```python
distortions = [] 
num_clusters = range(2, 7) 
for i in num_clusters: 
    centroids, distortion = kmeans(skala_data, i) 
    distortions.append(distortion) 
elbow_df = pd.DataFrame({'num_clusters': num_clusters, 'distortions': distortions})
sns.lineplot(x='num_clusters', y='distortions', data=elbow_df) 
plt.show()
```
